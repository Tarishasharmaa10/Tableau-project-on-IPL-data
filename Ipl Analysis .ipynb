## 1).Treating the City Column .

```python
ipl['city'].value_counts()
```

```python
ipl[ipl['city'].isnull()].shape
```

```python
ipl[ipl['city'].isnull()]
```

```python
ipl[ipl['venue']=='Sharjah Cricket Stadium']
```

```python
ipl[ipl['venue']=='Dubai International Cricket Stadium']
```

```python
ipl['city']=np.where(ipl['venue']=='Sharjah Cricket Stadium','Sharjah',ipl['city'])
```

```python
ipl['city']=np.where(ipl['venue']=='Dubai International Cricket Stadium','Dubai',ipl['city'])
```

```python
ipl['city'].isnull().sum()
```

## 2).Treat the Date Column .

```python
ipl['date']=pd.to_datetime(ipl['date'])  
```

```python
ipl.dtypes
```

### Feature Engineering related to date column .

```python
ipl['Season']=ipl['date'].dt.year
```

```python
ipl.head()
```

```python
ipl.dtypes
```

## 3).Treat the player of match column .

```python
ipl[ipl['player_of_match'].isnull()] 
```

```python
ipl['player_of_match']=ipl['player_of_match'].fillna('No player of match')
```

```python
ipl['player_of_match'].isnull().sum()
```

## 4).Treating the Winner column .

```python
ipl['winner']=ipl['winner'].fillna('No Winner')
```

## 5). Treating the Result column .

```python
ipl['result']=ipl['result'].fillna('No Result')
```

## 6). Treating the result margin column .

```python
ipl[ipl['result_margin'].isnull()]
```

```python
ipl['result_margin']=np.where(ipl['result']=='tie',0,ipl['result_margin'])
```

```python
ipl['result_margin']=np.where(ipl['result']=='No Result','Match Abanded',ipl['result_margin'])
```

## 7). Treating the eliminator column .

```python
ipl['eliminator'].value_counts()
```

```python
ipl[ipl['eliminator'].isnull()]
```

```python
ipl['eliminator'].fillna(ipl['eliminator'].mode()[0],inplace=True)
```

## 8).Treating the Method Column .

```python
ipl['method'].value_counts()
```

```python
ipl['method']=ipl['method'].fillna('No Method')
```
